```python
import random

# 단어 리스트 (과일 이름들)
words = ["apple", "banana", "orange", "grape", "lemon"]

def choose_word(words):
    """랜덤으로 단어 선택"""
    return random.choice(words)

def display_word(word, guessed_letters):
    """추측한 단어를 보여줌"""
    return ' '.join([letter if letter in guessed_letters else '_' for letter in word])

def display_hangman(attempts):
    """남은 기회에 따라 hangman 그림을 보여줌"""
    stages = [
        """
           -----
           |   |
               |
               |
               |
               |
        --------
        """,
        """
           -----
           |   |
           O   |
               |
               |
               |
        --------
        """,
        """
           -----
           |   |
           O   |
           |   |
               |
               |
        --------
        """,
        """
           -----
           |   |
           O   |
          /|   |
               |
               |
        --------
        """,
        """
           -----
           |   |
           O   |
          /|\  |
               |
               |
        --------
        """,
        """
           -----
           |   |
           O   |
          /|\  |
          /    |
               |
        --------
        """,
        """
           -----
           |   |
           O   |
          /|\  |
          / \  |
               |
        --------
        """
    ]
    return stages[6 - attempts]

def hangman():
    # 랜덤으로 단어 선택
    word = choose_word(words)
    guessed_letters = []
    attempts = 6  # 기회는 6번

    print("=== Hangman 게임 ===")
    print(display_word(word, guessed_letters))

    while attempts > 0:
        print(display_hangman(attempts))  # Hangman 그림 표시
        guess = input("\n추측할 문자를 입력하세요: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("한 개의 알파벳 문자를 입력하세요.")
            continue

        if guess in guessed_letters:
            print("이미 추측한 문자입니다.")
        elif guess in word:
            print(f"맞췄습니다! {guess}는 단어에 포함됩니다.")
            guessed_letters.append(guess)
        else:
            print(f"틀렸습니다! {guess}는 단어에 포함되지 않습니다.")
            guessed_letters.append(guess)
            attempts -= 1
            print(f"남은 기회: {attempts}")

        # 현재 상태 표시
        current_display = display_word(word, guessed_letters)
        print(current_display)

        # 단어를 다 맞췄을 때
        if '_' not in current_display:
            print("축하합니다! 단어를 맞췄습니다!")
            break
    else:
        print(display_hangman(attempts))  # 게임 종료 시 마지막 Hangman 그림 표시
        print(f"아쉽습니다! 정답은 {word}였습니다.")

# 게임 시작
hangman()
```
